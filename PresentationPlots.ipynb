## Plots for presentation

```python
# season
season_params = {
    "Winter": {"T_mu": 12.0, "T_sd": 5.0, "H_mu": 65.0, "H_sd": 10.0},
    "Spring": {"T_mu": 27.0, "T_sd": 6.0, "H_mu": 75.0, "H_sd": 12.0},
    "Summer": {"T_mu": 33.0, "T_sd": 5.0, "H_mu": 80.0, "H_sd": 14.0},
    "Fall":   {"T_mu": 28.0, "T_sd": 6.0, "H_mu": 75.0, "H_sd": 12.0},
}
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the x-axis ranges for each season
x = np.linspace(0, 50, 500)

# Plot the normal distributions for each season
plt.plot(x, norm.pdf(x, 12, 5), label='Winter (mu=12, sd=5)')
plt.plot(x, norm.pdf(x, 27, 6), label='Spring (mu=27, sd=6)')
plt.plot(x, norm.pdf(x, 33, 5), label='Summer (mu=33, sd=5)')
plt.plot(x, norm.pdf(x, 28, 6), label='Fall (mu=28, sd=6)')

# Add title, legend, and labels
plt.title('Normal Distributions for Different Seasons')
plt.xlabel('Temperature')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

```python
# Repeat for humidity
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the x-axis ranges for each season
x = np.linspace(40, 100, 1000)

# Plot the normal distributions for each season
plt.plot(x, norm.pdf(x, 65, 10), label='Winter (mu=65, sd=10)')
plt.plot(x, norm.pdf(x, 80, 14), label='Summer (mu=80, sd=14)')
plt.plot(x, norm.pdf(x, 75, 12), label='Fall & Spring (mu=75, sd=12)')

# Add title, legend, and labels
plt.title('Normal Distributions for Different Seasons')
plt.xlabel('Humidity')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```
